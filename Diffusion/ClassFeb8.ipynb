# 1D Random walk
## smaller title

$ \sum_1^100{1+i}$

```python
np.random.rand()
```

```python
stepsize = 1
numsteps = 100
position = []
x = 0
for i in range(numsteps):
    number = np.random.rand()
    if number < 0.5:
        x = x - stepsize
    else:
        x = x + stepsize
    position = position + [x]
    # could have typed
    # position.append(x)
    
```

```python
plt.plot(position)
```

```python
pos = np.array(position)
```

```python
np.average(pos)
```

```python
pos[-10]
```

```python
np.std(pos)
```

```python
df = pd.DataFrame(pos)
```

```python
df.plot()
```
